# Rosenblatt Perceptron

- Rosenblatt's perceptron is based on an approximation of how a neuron works:
     - linear combination of inputs
     - an output if the cell is "activated"

 0 & \mathrm{if} ~ \sum_{j} w_{j}x_{j} \leq ~ \mathrm{limiar} \\

\end{matrix}$$

## RadioGalaxyNET: Dataset

The RadioGalaxyNET dataset encompasses 2800 3-channel images featuring two radio sky channels and one corresponding infrared sky channel. Both noisy and processed radio images are included in the dataset. It comprises 2800 extended radio galaxies, resulting in a total of 4155 instances of these galaxies due to their proximity in the sky and their appearance in multiple images.

The radio galaxies in the dataset are categorized as:
- 13% FR-I <a name="cite_ref-1"></a>[<sup>[1]</sup>](#cite_note-1)
- 48% FR-II
- 14% FR-x
- 25% R sources

The dataset is divided into three sets: training, validation, and test sets, with a split ratio of  0.7:0.15:0.15

---
<a name="cite_note-1"></a>1. [Fanaroff–Riley classification](https://en.wikipedia.org/wiki/Fanaroff%E2%80%93Riley_classification)

## Annotations

The RadioGalaxyNET dataset has the following four sets of labels

- the infrared host galaxy positions.

- the extended radio galaxy classes,

- the bounding boxes enclosing all components of each radio galaxy, and

- the segmentation masks for radio galaxies.

Annotations in the RadioGalaxyNET dataset adhere to the COCO dataset format (Lin et al. Reference Lin, Fleet, Pajdla, Schiele and Tuytelaars2014), facilitating straightforward comparisons of **object detection methods**. Radio annotations for each galaxy are stored as ‘categories’, ‘box’, and ‘segmentation’, while the positions of the infrared hosts are stored as ‘keypoints’.

````python
import os

def list_files(startpath):
    for root, dirs, files in os.walk(startpath):
        level = root.replace(startpath, '').count(os.sep)
        indent = ' ' * 4 * (level)
        print('{}{}/'.format(indent, os.path.basename(root)))
        subindent = ' ' * 4 * (level + 1)
        for f in files:
            print('{}{}'.format(subindent, f))
````

````python
def calculate_bounding_boxes(image_shape, num_rows, num_cols):
    """
    Calculate bounding box coordinates for sectors in an image.
    
    Parameters:
        image_shape (tuple): Shape of the input image (height, width).
        num_rows (int): Number of rows for sectors.
        num_cols (int): Number of columns for sectors.
    
    Returns:
        list of numpy.ndarray: List containing bounding box coordinates for each sector.
    """
    height, width = image_shape
    sector_height = height // num_rows
    sector_width = width // num_cols

    bounding_boxes = []

    for row in range(num_rows):
        for col in range(num_cols):
            x1 = col * sector_width
            y1 = row * sector_height
            x2 = x1 + sector_width
            y2 = y1 + sector_height
            bounding_boxes.append(np.array([[x1, y1], [x2, y1], [x2, y2], [x1, y2]]))

    return bounding_boxes
````
